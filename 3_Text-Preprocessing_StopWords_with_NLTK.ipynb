### **POS Tagging :**

* Part of Speech (POS) tagging refers to assigning each word of a sentence to its part of speech.
* It is significant as it helps to give a better syntactic overview of a sentence.

```python
nltk.download('averaged-perceptron-tagger')
```

```python
stop_words = set(stopwords.words('english'))
snowballstemmer = SnowballStemmer('english')
```

```python
for i in range(len(sentences)):
  #sentences[i] = sentences[i].tolower()
  words = nltk.word_tokenize(sentences[i])
  words = [snowballstemmer.stem(word) for word in words if word.lower() not in stop_words]
  sentences[i] = ' '.join(words)
```
