```python
from pprint import pprint
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
myurl="https://stackoverflow.com/questions/43189302/syntaxerror-unexpected-eof-while-parsing" #specify the url
stk=urlopen(myurl).read()#query the website so that it returns a html page
soupified=BeautifulSoup(stk,'html.parser')#parse the html in the 'html' variable
```

```python
pprint(soupified.prettify()[:50000])
```

```python
#Converting html to text
text = soupified.prettify()[:50000]
```

```python
import nltk
nltk.download('punkt')
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
#Creating sentences
sentences = sent_tokenize(text)
print(len(sentences))
print(sentences)
```

```python
#Seggregating the words
words = word_tokenize(text)
print(len(words))
print(words)
```

```python
words_no_punc = []
for w in words:
  if w.isalnum():
    words_no_punc.append(w.lower())
    print(words_no_punc)
    print("\n")
    print("No. of words without punctuation = ",len(words_no_punc))
```

#Name Entity Recognition

```python
#nltk.download('all')
nltk.download('averaged_perceptron_tagger')
nltk.download('words')
nltk.download('maxent_ne_chunker')
#Sentence  for NER :
sentence="Mr.Smith made a deal on a beach of Switzerland near WHO."

#Tokenizing words :
tokenized_words = word_tokenize(sentence)

#POS Tagging :
for w in tokenized_words:
   tagged_words= nltk.pos_tag(tokenized_words)

print(tagged_words)
#Named Entity Recognition :
N_E_R = nltk.ne_chunk(tagged_words,binary=True)
print(N_E_R)

#To visualize :
#N_E_R.draw()
```
