## Wordnet Lemmatization
* Lemmatization technique is like stemming. The output will get after lemmatization is called 'lemma', which is a root word rather than root stem, the output of stemming.
* After lemmatization, we will be getting a valid word that means the same thing.
* NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example:  

```python
import nltk
```

```python
## Q&A, chatbots, text summarization
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```

```python
'''
POS - Noun-n
verb - v
adjective - a
adverb - r
'''
```

```python
words = ["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
lemmatizer.lemmatize("going",pos='a')
```

```python
lemmatizer.lemmatize("going",pos='v')
```

```python
lemmatizer.lemmatize("sportingly")
```

```python
for word in words:
  print(word + "---->" + lemmatizer.lemmatize(word, pos='n'))
```
